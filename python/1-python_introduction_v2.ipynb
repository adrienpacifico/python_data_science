## 1. Python : un langage interprété

Il existe deux grandes familles de langages de programmation :

| | **Langage compilé** (C, C++, Rust, Go) | **Langage interprété** (Python, R, JavaScript) |
|---|---|---|
| **Principe** | Le code source est traduit **en entier** en code machine *avant* l'exécution | Le code est lu et exécuté **ligne par ligne** par un programme appelé *interpréteur* |
| **Avantage** | Très rapide à l'exécution | Très rapide à itérer : on écrit, on exécute, on voit le résultat |
| **Inconvénient** | Cycle écriture → compilation → exécution plus long | Plus lent à l'exécution (mais souvent suffisant pour la data) |

Python est **interprété**. C'est pour ça qu'on peut l'utiliser de manière interactive — dans un terminal, dans un notebook, ou via un script.

## 3. Les commandes magiques d'IPython

Jupyter utilise **IPython** comme interpréteur. IPython ajoute des "commandes magiques" qui commencent par `%` (une ligne) ou `%%` (toute la cellule).

Voici les plus utiles au quotidien.

### `%%time` — mesurer le temps d'exécution d'une cellule

````python
%%time
# Combien de temps pour faire 10 millions de multiplications ?
total = 0
for i in range(10_000_000):
    total += i * 3
````

### `%%timeit` — mesurer le temps *moyen* (exécute plusieurs fois)

````python
%%timeit
# timeit exécute la cellule plusieurs fois et donne la moyenne
sum(range(100_000))
````

## 4. Le style en Python : PEP 8 et ruff

### Qu'est-ce qu'une PEP ?

Une **PEP** (*Python Enhancement Proposal*) est un document qui propose une amélioration ou une convention pour Python. C'est le processus officiel par lequel la communauté Python fait évoluer le langage.

Il existe des centaines de PEP, consultables sur [peps.python.org](https://peps.python.org/). Quelques exemples :

| PEP | Sujet |
|-----|-------|
| PEP 8 | Guide de style du code Python |
| PEP 20 | Le Zen de Python (`import this`) |
| PEP 484 | Les *type hints* (annotations de types) |
| PEP 723 | Metadata dans les scripts Python |

La plus importante pour le quotidien est la **PEP 8**.

### La PEP 8 — le guide de style officiel

La [**PEP 8**](https://peps.python.org/pep-0008/) définit les conventions de style pour le code Python. Pour que le code **fonctionne**, il suffit de respecter la syntaxe. Mais pour qu'il soit **lisible** par d'autres (et par vous-même dans 6 mois), il faut respecter un style cohérent.

Les règles principales :

- **Indentation** : 4 espaces (pas de tabulations)
- **Longueur de ligne** : max 79-120 caractères
- **Nommage** :
  - `ma_variable`, `ma_fonction` → snake_case (minuscules + underscores)
  - `MaClasse` → PascalCase
  - `MA_CONSTANTE` → MAJUSCULES
- **Espaces** : `x = 1` (pas `x=1`), `f(a, b)` (pas `f(a,b)`)
- **Imports** : un par ligne, en haut du fichier

#### Les conventions d'underscore

En Python, les underscores au début d'un nom ont une signification particulière :

| Convention | Signification | Exemple |
|---|---|---|
| `_ma_fonction` | **"Privé" par convention** — cette fonction/variable est un détail d'implémentation interne, elle n'est pas destinée à être utilisée par les utilisateurs du module ou de la classe. Rien ne l'interdit techniquement, mais c'est un signal : "ne comptez pas dessus, ça peut changer". | `_parse_raw_data()`, `_cache` |
| `__ma_methode` | **Name mangling** — Python renomme la méthode en interne (ex: `_MaClasse__ma_methode`) pour éviter les collisions dans l'héritage. Rarement utilisé en data science. | `__init_weights()` |
| `__init__` | **Méthode spéciale** (*dunder method*) — réservée à Python. Ce sont les méthodes "magiques" (`__init__`, `__len__`, `__str__`, etc.). On ne crée jamais ses propres noms en `__x__`. | `__init__()`, `__repr__()` |
| `variable_` | **Éviter un conflit** avec un mot-clé Python. | `class_`, `type_`, `id_` |

Le plus courant au quotidien : `_ma_fonction` pour signaler qu'une fonction est interne.

### ruff — un seul outil pour le style et la qualité

Historiquement, il fallait combiner plusieurs outils pour maintenir la qualité du code Python :
- **black** pour le formatage automatique
- **isort** pour trier les imports
- **flake8** pour détecter les erreurs de style
- **pylint** pour l'analyse statique

Depuis 2023, **[ruff](https://docs.astral.sh/ruff/)** remplace tous ces outils en un seul, et il est *extrêmement* rapide (écrit en Rust).

Deux commandes à retenir :

| Commande | Action |
|----------|--------|
| `ruff check fichier.py` | Vérifie le code (erreurs, imports inutilisés, style…) |
| `ruff format fichier.py` | Reformate automatiquement le code (comme black) |

## 5. Tester son code : pytest et le TDD

### Pourquoi tester ?

Quand on écrit du code, on peut facilement introduire des bugs sans s'en rendre compte — surtout quand le code grossit. Les **tests automatisés** sont des petits programmes qui vérifient que votre code fait bien ce qu'il est censé faire.

Pourquoi c'est important :
- **Confiance** : vous modifiez du code existant sans crainte de tout casser
- **Documentation vivante** : les tests montrent *comment* le code est censé fonctionner
- **Collaboration** : un collègue (ou un LLM) peut modifier votre code, les tests valident que rien n'est cassé
- **Avec les LLM** : vous demandez à un LLM de générer du code → les tests permettent de vérifier automatiquement qu'il fait bien ce que vous attendez

### pytest — l'outil standard pour tester en Python

**[pytest](https://docs.pytest.org/)** est le framework de test le plus utilisé en Python. Son principe est simple : on écrit des fonctions qui commencent par `test_` et qui utilisent `assert` pour vérifier des conditions.

````python
%%writefile /tmp/calculs.py
"""Module avec quelques fonctions de calcul."""


def celsius_vers_fahrenheit(celsius):
    """Convertit une température de Celsius vers Fahrenheit."""
    return celsius * 9 / 5 + 32


def moyenne(nombres):
    """Calcule la moyenne d'une liste de nombres."""
    if not nombres:
        raise ValueError("La liste ne peut pas être vide")
    return sum(nombres) / len(nombres)
````

````python
%%writefile /tmp/test_calculs.py
"""Tests pour le module calculs."""
import pytest
from calculs import celsius_vers_fahrenheit, moyenne


def test_celsius_vers_fahrenheit_eau_gele():
    assert celsius_vers_fahrenheit(0) == 32


def test_celsius_vers_fahrenheit_eau_bout():
    assert celsius_vers_fahrenheit(100) == 212


def test_moyenne_simple():
    assert moyenne([10, 20, 30]) == 20


def test_moyenne_un_seul_element():
    assert moyenne([42]) == 42


def test_moyenne_liste_vide_leve_erreur():
    with pytest.raises(ValueError):
        moyenne([])
````

````python
# Lancer les tests avec pytest
!cd /tmp && python -m pytest test_calculs.py -v
````

### Le TDD (Test-Driven Development) en bref

Le TDD est une méthode de développement où on **écrit les tests avant le code** :

1. **Red** : écrire un test qui échoue (le code n'existe pas encore)
2. **Green** : écrire le minimum de code pour que le test passe
3. **Refactor** : améliorer le code tout en gardant les tests verts

C'est particulièrement puissant avec les LLM : vous écrivez les tests (qui décrivent *ce que vous voulez*), puis vous demandez au LLM de générer le code qui les fait passer. Les tests deviennent votre **spécification** et votre **filet de sécurité**.
